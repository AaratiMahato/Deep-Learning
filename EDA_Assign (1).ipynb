```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('/content/it_jobs_2030.csv')
```

```python
# Display basic information about the dataset
print("Basic Information:")
print(data.info())
print("\nFirst few rows of the dataset:")
print(data.head())
```

```python

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:")
print(data.duplicated().sum())
```

```python
# Visualize the distribution of numerical variables
numerical_columns = data.select_dtypes(include=['number']).columns  # Include 'number' to capture all numeric types

if len(numerical_columns) > 0:  # Proceed only if numerical columns exist
    for col in numerical_columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(data[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.show()
else:
    print("No numerical columns found for visualization.")
```

```python
# Visualize the distribution of categorical variables
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Check if numerical_columns is not empty before creating the heatmap
if len(numerical_columns) > 0:
    plt.figure(figsize=(12, 8))
    sns.heatmap(data[numerical_columns].corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("No numerical columns found for correlation heatmap.")
```

```python
if len(numerical_columns) > 0:  # Proceed only if numerical columns exist
    sns.pairplot(data[numerical_columns])
    plt.show()
else:
    print("No numerical columns found for pairplot.")
```
